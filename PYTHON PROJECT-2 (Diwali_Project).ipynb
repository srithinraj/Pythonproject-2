Exploratory Data Analysis
=====================


Gender

```python
sns.countplot(x='Gender',data=df,color='Green')
```

```python
sales_gen=df.groupby("Gender")['Amount'].sum().sort_values()
```

```python
#Agewise and genderwise classification of custoers
sns.countplot(x='Age Group',data=df,hue='Gender')
```

Womens in the Age between 26-35 is the majority customers

```python
sales_gen
```

State



```python
#Statewise orders
state_orders=df.groupby("State")['Orders'].sum().sort_values().plot(kind="barh")
```

```python
Product_wise=df.groupby('Product_Category')['Amount'].sum()
```

```python
Product_wise.plot(kind='barh')
plt.xlabel('Product_Category')
plt.ylabel('Amount')
plt.title('Product vs Sales')
```

Conclusion:  Women in the age group of 26-35 working inthe IT , Healthcare and Aviation from UP, Maharashtra and Karnataka more likely to buy products from food ,clothing and electronics category
